```python
import cv2
src = cv2.imread(r'C:\Users\Administrator\Pictures\Screenshots\Screenshot 2026-02-11 213943.png')  # Read the image

# Convert to Grayscale
gray_image = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

# Display
cv2.imshow("Grayscale Image", gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
src = cv2.imread(r'C:\Users\Administrator\Pictures\Screenshots\Screenshot 2026-02-11 213943.png')

# Convert to HSV
hsv_image = cv2.cvtColor(src, cv2.COLOR_BGR2HSV)

cv2.imshow("HSV Image", hsv_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt
src = cv2.imread(r'C:\Users\Administrator\Pictures\Screenshots\Screenshot 2026-02-11 213943.png')

# Convert from BGR to RGB
rgb_image = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)

# Display with Matplotlib
plt.imshow(rgb_image)
plt.title("RGB Image for Matplotlib")
plt.axis('off')
plt.show()
```
